# Data Manipulation

```python
def inplant(series):
    return series.fillna(series.median())
def inplant2(series):
    return series.fillna(series.mean())
def inplant3(series):
    return series.fillna(series.median())
```

```python
cardio[73:80]
```

```python
cardio[12:20]
```

```python
cardio[30:33]
```

```python
cardio['Education']=cardio['Education'].transform(inplant)
cardio['Age']=cardio['Age'].transform(inplant2).apply(lambda x: int(x))
cardio['Income']=cardio['Income'].transform(inplant3)
```

# Data Visulization

```python
'''
1.I want to check Product Usage
2.Age releationship with Miles // matlab kitna age ka person kitna miles chalta hai
3.product buyers income
'''
```

```python
use=cardio.groupby(cardio['Product']).Usage.agg(np.median)
mile_age = cardio.groupby(cardio['Age']).Miles.agg(np.mean)
income=cardio.groupby(cardio['Product']).Income.agg(np.sum)
```

```python
plt.figure(figsize=(16,5))
plt.plot(use,'ro',color='b')
plt.title('Product Usage')
plt.xlabel('Product')
plt.ylabel('Usage')
```

```python
plt.figure(figsize=(16,5))
plt.plot(mile_age,color='g')
plt.title('Age decides Miles')
plt.xlabel('Age')
plt.ylabel('Miles')
```

```python
plt.figure(figsize=(16,5))
plt.plot(income,'ro',color='r')
plt.title('Product Buyer\'s Income')
plt.xlabel('Product')
plt.ylabel('Income')
```
